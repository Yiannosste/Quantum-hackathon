# 02 — Simulation Phase: develop against the local emulator

The whole pipeline in one call. Everything you see here works **unchanged** on real hardware later — only the config file changes.

```python
import sys; sys.path.insert(0, "../..")
from problems.vrp import VehicleRoutingProblem
from qlkit import LogisticsSolver, GreedyAssignmentRepair

problem = VehicleRoutingProblem.small_instance()
solver = LogisticsSolver.from_config(problem, "../../configs/local.json",
                                     repair=GreedyAssignmentRepair())

result = solver.solve(lambdas={"one_hot": 12.0, "capacity": 2.0})
print(f"feasible: {result.samples.feasible_fraction:.1%}  "
      f"repaired: {result.samples.repaired_fraction:.1%}  "
      f"best cost: {result.metrics.true_objective:.2f}")
result.best_solution.payload["routes"]
```

```python
# Inspect the normalized SampleSet — every record is decoded and scored:
for record in sorted(result.samples.records, key=lambda r: r.energy)[:5]:
    print(f"count={record.count:4d}  energy={record.energy:8.2f}  "
          f"true_cost={record.true_objective:6.2f}  feasible={record.hard_feasible}  "
          f"violations={record.violations}")
```

```python
# Validation Phase preview: this is EXACTLY the code the judges run.
print(solver.validate(result.best_solution))
solver.close()
```
